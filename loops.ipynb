```python
#ðŸŸ¨ Find First Engineer (1.13.4)
job_titles = ['Data Scientist', 'Data Analyst', 'Machine Learning Engineer', 'Software Engineer',  'Data Engineer']

index=0
count =0

while index < len(job_titles):
    if 'Engineer' in  job_titles[index]:
        print(job_titles[index])
        count += 1
        if count == 2:   # stop after 2nd occurrence
            break
    index+=1
    
```
