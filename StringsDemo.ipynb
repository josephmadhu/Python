```python
#Convert to Uppercase (1.4.1)
job_title = 'data scientist'
job_title.upper()
job_title = job_title.upper()
print(job_title)
```

```python
#Replace Substring (1.4.2) 
job_title = 'Data Analyst'
job_title= job_title.replace( 'Analyst' , 'Engineer' )
#help(str)
print(job_title)
```

```python
#Concatenate Strings (1.4.3)
seniority_level = 'Senior'
job_title = 'Data Analyst'
print(job_title + " " + seniority_level)
print("__add__ accepts only two arugument self and other: ", str.__add__(job_title,seniority_level))
```

```python
#Check String Length (1.4.4)
job_title = 'data scientist'
print("Job title Length:", len(job_title))
print("Job title Length:", job_title.__len__())
```

```python
#Convert to Lowercase (1.4.5)
job_title = 'data scientist'.upper()
print(job_title)
job_title.lower()
job_title = job_title.lower()
print(job_title)
print(job_title.capitalize())
```
