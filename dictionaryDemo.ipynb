Problem Statement:

Access the value of the number of job postings for the key 'Data Analyst' from the dictionary job_postings and print it. The dictonary job_postings is set to {'Data Scientist': 120, 'Data Analyst': 80, 'Machine Learning Engineer': 50}.


```python
job_postings = {'Data Scientist': 120, 'Data Analyst': 80, 'Machine Learning Engineer': 50}
print(job_postings['Data Analyst'])
```

```python
#ðŸŸ© Add Job Role (1.9.2) - Problem
job_postings = {'Data Scientist': 120, 'Data Analyst': 80, 'Machine Learning Engineer': 50}
job_postings.update({'AI Specialist' : 30, 'Data Enginer': 50})
print(job_postings)
#another waay
job_postings['AI Engineer' ] =90
print(job_postings)
print(job_postings['Machine Learning Engineer'])

#converting to list and accessing based on index value
print (list(job_postings.keys())[2])
print(list(job_postings.values())[2])
print(list(job_postings.items())[2])
```
