```python
#Concatenate Strings (1.5.1)
job_title = 'Data Scientist'
print("Job Title :", job_title)
```

```python
#Format with f-strings (1.5.2) 
position = 'Data Scientist'
company = 'DataWiz Inc.'
print("1. State: {} at {}".format(position, company))
print(f'2. State: {position} at {company}')
print('3. State: ' + position + ' at ' + company)
```

```python
#Format with str.format() (1.5.3)
role = 'Data Analyst'
skill = 'Python'
finalrole = "Role: {}, Skill: {}".format(role,skill) 
print(finalrole)
```

```python
#Format with %-formatting (1.5.4)
role = 'Data Engineer'
skill = 'SQL'
result = 'Role: %s, Skill: %s' % (role, skill)
testskill = 'test %s' % skill
print(result)
print(testskill)
```
